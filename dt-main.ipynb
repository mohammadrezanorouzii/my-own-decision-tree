```python
def train_test_split(df, test_size):
    
    test_indices = random.sample(population=df.index.tolist(), k=int(test_size*df.shape[0]))
    test_df = df.loc[test_indices]
    train_df = df.drop(test_indices)
    
    return train_df, test_df

train_df, test_df = train_test_split(df, test_size=0.2)
train_df.shape
```

```python
def node_is_pure(data):
    if len(np.unique(data[:, -1], return_counts=False)) == 1:
        return True
    else:
        return False
```

```python
def classify_node(data):
    
    unique, counts = np.unique(data[:,-1], return_counts=True)
    index = counts.argmax()
    node_class = unique[index]
    
    return node_class
```

```python
def split(data, column, split_point):

    below = data[:split_point, column]
    above = data[split_point:, column]

    return above, below
```

```python
def every_possible_splits(data) :

    every_splits = {}
    for column in range(data.shape[1]):
        every_splits[column] = []
        all_values = np.unique(data[:, column])

        for i in range(len(all_values)-1):
            every_splits[column].append((all_values[i] + all_values[i+1]) / 2)

    return every_splits
```
